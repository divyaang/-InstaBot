## 1.1 Login to your Instagram Handle

```python
def login(id, password):
    
    wait = WebDriverWait(driver, 10)
    text_box = wait.until(EC.presence_of_all_elements_located((By.XPATH, '//input[@class = "_2hvTZ pexuQ zyHYP"]')))
    
    text_box[0].send_keys(id)
    text_box[1].send_keys(password)  
    
    login_button = driver.find_element_by_xpath('//button[@class = "sqdOP  L3NKy   y3zKF     "]')
    login_button.submit()
```

## 2.1 Type for “food” in search bar and print all the names of the Instagram Handles that are displayed in list after typing “food”

```python
def print_list(word):

    driver.refresh()
    wait = WebDriverWait(driver, 10)
    search_box = wait.until(EC.presence_of_element_located((By.XPATH, '//input[contains(@class, "XTCLo x3qfX")]')))

    search_box.clear()
    search_box.send_keys(word)
    
    _list = wait.until(EC.presence_of_all_elements_located((By.XPATH, '//a[contains(@class, "yCE8d")]')))
    
    data = driver.page_source
    html_data = BeautifulSoup(data, 'html.parser')
    
    hlist = html_data.find_all(class_ = 'Ap253')

    print('The names of all the Instagram Handles that are displayed in list after typing', word, 'are:')

    for handle in hlist:
        name = handle.string
        if name[0] == '#':
            print(name[1:])
        else:
            print(handle.string)
```

## 4.2 Follow/Unfollow given handle - Start following it. Print a message if you are already following

```python
def check_follow():
    
    driver.refresh
    wait = WebDriverWait(driver, 10)
    follow_button = wait.until(EC.element_to_be_clickable((By.XPATH, '//button[contains(@class, "_5f5mN")]')))

    data = driver.page_source
    html_data = BeautifulSoup(data, 'html.parser')

    follow_unfollow = html_data.find(class_ = '_5f5mN').string

    if follow_unfollow == 'Follow':
        print('You were not following So Delhi')
        follow_button.click()
        print('Now, you have started following So Delhi')

    else:
        print('You are already following So Delhi')
```

## 5.1 Like/Unlike posts - Liking the top 30 posts of the ‘dilsefoodie'. Print message if you have already liked it.

```python
from selenium.common.exceptions import ElementNotInteractableException

def like_30():
    
    wait = WebDriverWait(driver, 10)
    i = 0
    all_posts = []

    try:
        while len(all_posts) < 30:

            new_post = wait.until(EC.presence_of_all_elements_located((By.XPATH, '//div[contains(@class, "v1Nh3 kIKUG  _bz0w")]/a')))

            for p in new_post[len(all_posts):]:
                p.click()

                like_button = wait.until(EC.element_to_be_clickable((By.XPATH, '//span[contains(@class, "fr66n")]')))
                like_button.click()

                exit_button = wait.until(EC.element_to_be_clickable((By.XPATH, '//div[contains(@class, "                    Igw0E     IwRSH      eGOV_         _4EzTm                                                                                  BI4qX            qJPeX            fm1AK   TxciK yiMZG")]')))
                exit_button.click()

                i=i+1
                print("Post",i,"liked")
                if i==30:
                    break

            all_posts = new_post[len(all_posts):] + all_posts
            current_height = driver.execute_script('return document.body.scrollHeight;')
            driver.execute_script('window.scrollTo(0, arguments[0]);', current_height) 

    except ElementNotInteractableException:
        pass
    
    
```

## 6.1: Extract list of followers - Extract the usernames of the first 500 followers of ‘foodtalkindia’ and ‘sodelhi’

```python
def get_username(profile):
    search_and_open(profile)
    
    wait = WebDriverWait(driver, 10)

    followers_button =  wait.until(EC.element_to_be_clickable((By.XPATH, '//a[contains(@class, "-nal3")]')))
    followers_button.click()
    
    userlist = []

    while len(userlist) < 500:

        new_posts = wait.until(EC.presence_of_all_elements_located((By.XPATH, '//a[contains(@class, "FPmhX notranslate  _0imsa ")]')))

        data = driver.page_source
        html_data = BeautifulSoup(data, 'html.parser')

        ul = html_data.find_all(class_ = 'd7ByH')
        for u in ul[len(userlist):]:
            userlist.append(u.a.string)

        for list in new_posts:
            driver.execute_script("arguments[0].scrollIntoView();", list)
        
    return userlist
```

## 7: Check the story of ‘coding.ninjas’. Consider the following Scenarios and print error messages accordingly -
#### If You have already seen the story.
#### Or The user has no story.
#### Or View the story if not yet seen

```python
from selenium.common.exceptions import TimeoutException

def check_story(profile):
    search_and_open(profile)
    wait = WebDriverWait(driver, 10)
    time.sleep(5)
    
    try:
        story = wait.until(EC.element_to_be_clickable((By.XPATH, '//div[contains(@class, "RR-M- h5uC0")]')))
        
        data = driver.page_source
        html_data = BeautifulSoup(data, 'html.parser')

        c = html_data.find(class_ = 'CfWVH')
        h = int(c['height'])
        w = int(c['width'])

        if h == 166 and w == 166:
            print("You have already seen the story")

        if h == 168 and w == 168:
            print("You have not seen the story")
            story.click()
            print("You are now seening the story")

    except TimeoutException:
        print("The user has no story")
```

```python
check_story('coding.ninjas')
```
